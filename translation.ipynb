# Tweets from the earthqauke

Preporcessing

```python
import re
import string


def clean_text(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text) # remove mentions
    text = re.sub(r'#', '', text) # remove hashtags
    text = re.sub(r'\s+', ' ', text) # remove extra whitespace
    return text
```

```python
data["content"] = data["content"].apply(clean_text)
data.head()
```

## Set up translation service

```python
import os
from google.cloud import translate_v2 as translate
import os


os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = "api.json"

translate_client = translate.Client()
df = pd.read_csv('original_dataframe.csv')

# Define a function to translate a single text
def translate_text(text, target_language):
    result = translate_client.translate(text, target_language=target_language)
    return result['translatedText']

# Translate the "content" column of the dataframe
df['translated_content'] = df['content'].apply(lambda x: translate_text(x, 'fr'))

# Save the translated dataframe to a new CSV file
df.to_csv('translated_dataframe.csv', index=False)
df.head()
```
